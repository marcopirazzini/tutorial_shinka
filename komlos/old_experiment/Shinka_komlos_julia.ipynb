# n=12

```julia
function pmiter(n)
    s0 =  -1:2:1
    s = Base.Iterators.product([s0 for i in 1:n]...)
    s1 = Base.Iterators.map(x -> [x...], s)
end
    
function komlos_mapin(A)
    for i in 1:size(A,2)
        A[:,i] .= A[:,i] ./ norm(A[:,i])
    end
    return A
end

komlos(A) = minimum([maximum(abs.(A*v)) for v in pmiter(size(A,2))])
```

```julia
A = readdlm("shinka_komlos_n12.txt")
```

```julia
komlos(komlos_mapin(A))
```

```julia
Ar = round.(A,digits=4)
```

```julia
komlos(komlos_mapin(Ar))
```
