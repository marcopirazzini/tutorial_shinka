### Standardizing the data + NN prep

```python
plotted_tracks_df.head()
```

```python
astec_grid = plotted_tracks_df.copy()

astec_grid = astec_grid[['M', 'Z', 'star_age', 'Teff', 'luminosity', 'delta_nu', 'nu_max', 'Track']]

astec_grid['Teff'] = np.log10(astec_grid['Teff'])
astec_grid['luminosity'] = np.log10(astec_grid['luminosity'])
astec_grid['delta_nu'] = np.log10(astec_grid['delta_nu'])
astec_grid['nu_max'] = np.log10(astec_grid['nu_max'])
```

```python
astec_grid.head()
```

```python
def standardize_stellar_data(df, features=['Teff', 'luminosity', 'delta_nu', 'nu_max']):
    df_std = df.copy()
    stats = {}
    
    for feat in features:
        mean = df[feat].mean()
        std = df[feat].std()
        df_std[feat] = (df[feat] - mean) / std
        stats[feat] = {'mean': mean, 'std': std}
        
    return df_std, stats
```

```python
def unstandardize(data, stats, feature_names):
    unstd_data = data.copy()
    for i, name in enumerate(feature_names):
        mean = stats[name]['mean']
        std = stats[name]['std']
        unstd_data[:, i] = (data[:, i] * std) + mean
    return unstd_data
```

```python
def split_by_track(df, train_size=0.7, val_size=0.15, test_size=0.15):
    unique_tracks = df['Track'].unique()
    
    train_tracks, test_val_tracks = train_test_split(
        unique_tracks, train_size=train_size, random_state=42
    )
    
    relative_val_size = val_size / (val_size + test_size)
    val_tracks, test_tracks = train_test_split(
        test_val_tracks, train_size=relative_val_size, random_state=42
    )
    
    train_df = df[df['Track'].isin(train_tracks)]
    val_df = df[df['Track'].isin(val_tracks)]
    test_df = df[df['Track'].isin(test_tracks)]
    
    return train_df, val_df, test_df
```

```python
features = ['M', 'Z', 'star_age'] 

targets = ['Teff', 'luminosity', 'delta_nu', 'nu_max']

df_std_features, feature_stats = standardize_stellar_data(astec_grid, features=features)

df_std_all, target_stats = standardize_stellar_data(df_std_features, features=targets)

train_df, val_df, test_df = split_by_track(df_std_all)
```

### Building simple FLAX network

```python
import jax
import jax.numpy as jnp
import flax.linen as nn
from flax.training import train_state
import optax
```

```python
class StellarNet(nn.Module):
    @nn.compact
    def __call__(self, x):
        x = nn.Dense(128)(x)
        x = nn.relu(x)
        x = nn.Dense(128)(x)
        x = nn.relu(x)
        # Output 4 values: Teff, L, delta_nu, nu_max
        x = nn.Dense(4)(x) 
        return x

def run_stellar_nn_cpu(train_df, val_df, test_df, features, target, epochs=100, batch_size=1024, print_every=50):
    X_train = jnp.array(train_df[features].values, dtype=jnp.float32)
    y_train = jnp.array(train_df[target].values, dtype=jnp.float32)
    X_val = jnp.array(val_df[features].values, dtype=jnp.float32)
    y_val = jnp.array(val_df[target].values, dtype=jnp.float32)
    num_samples = X_train.shape[0]

    model = StellarNet()
    key = jax.random.PRNGKey(0)
    params = model.init(key, X_train[:1])['params']
    
    tx = optax.adam(learning_rate=0.001)
    state = train_state.TrainState.create(apply_fn=model.apply, params=params, tx=tx)

    total_params = sum(x.size for x in jax.tree_util.tree_leaves(params))
    trainable_params = sum(x.size for x in jax.tree_util.tree_leaves(state.params))
    print(f"Total parameters: {total_params:,}")

    @jax.jit
    def train_step(state, batch_x, batch_y):
        def loss_fn(params):
            pred = state.apply_fn({'params': params}, batch_x)
            return jnp.mean((pred - batch_y) ** 2)
        loss, grads = jax.value_and_grad(loss_fn)(state.params)
        state = state.apply_gradients(grads=grads)
        return state, loss

    @jax.jit
    def calculate_loss(params, x, y):
        pred = model.apply({'params': params}, x)
        return jnp.mean((pred - y) ** 2)

    print(f"Training on {num_samples} rows using {jax.device_count()} CPU devices...")
    start_time = time.time()
    interval_start = start_time

    for epoch in range(1, epochs + 1):
        key, subkey = jax.random.split(key)
        perms = jax.random.permutation(subkey, num_samples)
        
        epoch_loss = 0.0
        for i in range(0, num_samples, batch_size):
            batch_idx = perms[i : i + batch_size]
            state, loss = train_step(state, X_train[batch_idx], y_train[batch_idx])
            epoch_loss += loss
            
        if epoch % print_every == 0:
            # Synchronize for timing and accuracy
            jax.block_until_ready(state.params)
            
            # Calculate validation loss on the full val set
            v_loss = calculate_loss(state.params, X_val, y_val)
            # Average training loss over the last epoch's batches
            avg_train_loss = epoch_loss / (num_samples // batch_size)
            
            current_time = time.time()
            interval_duration = current_time - interval_start
            
            print(f"Epoch {epoch:4d} | Train Loss: {avg_train_loss:.6f} | Val Loss: {v_loss:.6f} | Time: {interval_duration:.2f}s")
            interval_start = current_time

    X_test = jnp.array(test_df[features].values, dtype=jnp.float32)
    y_test = jnp.array(test_df[target].values, dtype=jnp.float32)
    predictions = state.apply_fn({'params': state.params}, X_test)
    residuals = np.array(y_test) - np.array(predictions)
    
    return predictions, residuals
```

### Running FLAX

```python
predictions, residuals = run_stellar_nn_cpu(train_df, val_df, test_df, features, target=targets, epochs=500)
```

### Viewing Results

```python
actual_phys = unstandardize(np.array(test_df[targets]), target_stats, targets)
pred_phys = unstandardize(np.array(predictions), target_stats, targets)

inputs_phys = unstandardize(np.array(test_df[features]), feature_stats, features)
```

```python
actual_phys_unlogged = 10**actual_phys
pred_phys_unlogged = 10**pred_phys
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for i, name in enumerate(targets):
    ax = axes[i]
    ax.scatter(actual_phys_unlogged[:, i], pred_phys_unlogged[:, i], alpha=0.3, s=1, color='teal')
    
    # Draw the perfect prediction line
    lims = [np.min([ax.get_xlim(), ax.get_ylim()]), 
            np.max([ax.get_xlim(), ax.get_ylim()])]
    ax.plot(lims, lims, 'r--', alpha=0.75, zorder=3)
    
    ax.set_title(f'Target: {name}')
    ax.set_xlabel('Actual Value')
    ax.set_ylabel('Predicted Value')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for i, name in enumerate(targets):
    ax = axes[i]
    # Calculate percentage error: (Pred - Actual) / Actual * 100
    pct_error = (pred_phys_unlogged[:, i] - actual_phys_unlogged[:, i]) / actual_phys_unlogged[:, i] * 100
    
    ax.hist(pct_error, bins=50, color='royalblue', edgecolor='black', alpha=0.7)
    ax.axvline(0, color='red', linestyle='--')
    
    ax.set_title(f'Error Distribution: {name}')
    ax.set_xlabel('Percentage Error (%)')
    ax.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
